```python
import matplotlib.pyplot as plt
import numpy as np
y = np.array([35,50,15])
plt.pie(y)
plt.show()
```

# pie chart with labels

```python
import matplotlib.pyplot as plt
import numpy as np
y = np.array([35,50,15])
co=["r","b","hotpink"]
plt.pie(y,colors=co,labels=["DSP","CD","WT"])
plt.pie(y)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
y = np.array([35,50,15])
co=["r","b","hotpink"]
plt.pie(y,colors=co,labels=["DSP","CD","WT"])
plt.pie(y)
plt.legend()
plt.show()
```

# pie chart with title

```python
import matplotlib.pyplot as plt
import numpy as np
y = np.array([35,50,15])
co=["r","b","hotpink"]
plt.pie(y,colors=co,labels=["DSP","CD","WT"])
plt.pie(y)
plt.legend(title="subjects")
plt.show()
```
