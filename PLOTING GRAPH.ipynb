```python
import numpy as np
from matplotlib import pyplot as plt
x=np.arange(1,11)
y=2 * x + 5
plt.title("line plot")
plt.xlabel("x axis caption")
plt.ylabel("y axis caption")
plt.plot(x,y)
plt.show()
```

```python
plt.plot([5,100],[5,200],"--")
plt.show()
```

```python
x=[1,2,3,4,5]
y=[10,20,30,40,50]
plt.scatter(x,y)
plt.show()
```
