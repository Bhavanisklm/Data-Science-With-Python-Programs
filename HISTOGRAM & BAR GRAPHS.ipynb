# histograms

```python
import matplotlib.pyplot as plt
numbers=[0.1,0.5,1.5,2.4,5.5,6,8,9]
plt.hist(numbers)
plt.xlabel("numbers")
plt.ylabel("frequency")
plt.show()
```

```python
import matplotlib.pyplot as plt
numbers=[0.1,0.5,1.5,2.4,5.5,6,8,9]
plt.hist(numbers,bins=3,color="r")
#plt.hist(numbers,bins=10,color="r")
plt.xlabel("numbers")
plt.ylabel("frequency")
plt.show()
```

# Bar graphs

```python
import matplotlib.pyplot as plt
country=["USA","UK","INDIA","FRANCE","GERMANY"]
GDP_per_capita=[48000,44000,50000,45000,38000]
plt.bar(country,GDP_per_capita)
plt.title("country vs GDP_per_capita")
plt.xlabel("country")
plt.ylabel("GDP_per_capita")
plt.show()
```

```python
import matplotlib.pyplot as plt
country=["USA","UK","INDIA","FRANCE","GERMANY"]
GDP_per_capita=[48000,44000,50000,45000,38000]
plt.bar(country,GDP_per_capita,color="pink")
plt.title("country vs GDP_per_capita")
plt.xlabel("country")
plt.ylabel("GDP_per_capita")
plt.show()
```
